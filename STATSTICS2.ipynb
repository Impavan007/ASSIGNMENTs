```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Generate a discrete uniform distribution
low, high = 1, 6  # Simulating a six-sided die
data = randint.rvs(low, high+1, size=10000)

# Calculate PMF
values, counts = np.unique(data, return_counts=True)
pmf = counts / len(data)

# Plot the PMF
plt.bar(values, pmf, color='skyblue', edgecolor='black')
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

```python
def bernoulli_pdf(x, p):
    """Calculate the PDF of a Bernoulli distribution."""
    if x not in [0, 1]:
        return 0
    return p if x == 1 else 1 - p

# Test the function
p = 0.7  # Probability of success
print(f"PDF for x=1: {bernoulli_pdf(1, p)}")
print(f"PDF for x=0: {bernoulli_pdf(0, p)}")
```

```python
# Parameters
low, high = 1, 6
data = randint.rvs(low, high+1, size=10000)

# Calculate CDF
values, counts = np.unique(data, return_counts=True)
pmf = counts / len(data)
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(values, cdf, where='post', color='blue')
plt.title('CDF of a Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.show()
```

```python
def calculate_z_scores(data):
    """Calculate Z-scores for a dataset."""
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

# Generate random data
data = np.random.normal(0, 1, 100)
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.hist(z_scores, bins=20, color='orange', edgecolor='black')
plt.title('Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Simulate a non-normal distribution (e.g., exponential)
data = np.random.exponential(scale=2, size=10000)

# Take samples of increasing size and calculate sample means
sample_means = []
sample_sizes = [5, 10, 30, 50, 100]

for size in sample_sizes:
    means = [np.mean(np.random.choice(data, size, replace=True)) for _ in range(1000)]
    sample_means.append(means)

# Plot the histograms of sample means
fig, axes = plt.subplots(1, len(sample_sizes), figsize=(20, 4))
for i, size in enumerate(sample_sizes):
    axes[i].hist(sample_means[i], bins=30, density=True, color='lightblue', edgecolor='black')
    axes[i].set_title(f'Sample Size {size}')
    axes[i].set_xlabel('Sample Mean')
    axes[i].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate multiple samples from a normal distribution
mean, std_dev = 10, 2
population = np.random.normal(mean, std_dev, 10000)

# Verify the Central Limit Theorem
sample_means = []
sample_sizes = [5, 10, 30, 50, 100]

for size in sample_sizes:
    means = [np.mean(np.random.choice(population, size, replace=True)) for _ in range(1000)]
    sample_means.append(means)

# Plot the histograms of sample means
fig, axes = plt.subplots(1, len(sample_sizes), figsize=(20, 4))
for i, size in enumerate(sample_sizes):
    axes[i].hist(sample_means[i], bins=30, density=True, color='lightgreen', edgecolor='black')
    axes[i].set_title(f'Sample Size {size}')
    axes[i].set_xlabel('Sample Mean')
    axes[i].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import norm

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, loc=0, scale=1)

    plt.plot(x, pdf, label='Standard Normal Distribution', color='blue')
    plt.title('Standard Normal Distribution (mean=0, std=1)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid()
    plt.legend()
    plt.show()

plot_standard_normal()
```

```python
def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev

# Example data
data_point = 15
mean, std_dev = 10, 2
z_score = calculate_z_score(data_point, mean, std_dev)

print(f"Z-Score: {z_score}")
print(f"Probability from Standard Normal Distribution: {norm.cdf(z_score):.4f}")
```

```python
def hypothesis_testing(sample_mean, population_mean, std_dev, n, alpha=0.05):
    z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
    critical_value = norm.ppf(1 - alpha / 2)
    return z, critical_value

# Example data
z, critical_value = hypothesis_testing(12, 10, 2, 30)
print(f"Z-Statistic: {z}, Critical Value: {critical_value}")
```

```python
from scipy.stats import sem, t

# Example dataset
data = np.random.normal(10, 2, 30)

# Confidence interval
confidence = 0.95
mean = np.mean(data)
margin_of_error = t.ppf((1 + confidence) / 2, len(data) - 1) * sem(data)
ci = (mean - margin_of_error, mean + margin_of_error)

print(f"Confidence Interval: {ci}")
```

```python
data = np.random.normal(10, 2, 30)
ci = (np.mean(data) - 1.96 * sem(data), np.mean(data) + 1.96 * sem(data))
print(f"Confidence Interval for Mean: {ci}")
```
